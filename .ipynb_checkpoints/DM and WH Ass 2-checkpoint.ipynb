# Checking Outliers

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def visualize_outliers(df, columns):    
    for col in columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[col], color="lightblue")
        plt.title(f'Boxplot of {col}')
        plt.show()
```

```python
columns_to_check = [
    "Discount", 
    "Profit", 
    "Quantity", 
    "Sales", 
    "DaystoShipActual", 
    "SalesForecast",
    "DaystoShipScheduled",
    "SalesperCustomer",
    "ProfitRatio",
    "latitude",
    "longitude"
]


visualize_outliers(df, columns_to_check)
```

# Checking Skewness

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_skewness(df, columns):
    for col in columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[col], kde=True, bins=30, color='skyblue')
        plt.title(f'Histogram and KDE of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()
```

```python
plot_skewness(df, columns_to_check)
```

# Feature Engineering

```python
def feature_eng(df):
    # Ensure date columns are in datetime format
    df['OrderDate'] = pd.to_datetime(df['OrderDate'])
    df['ShipDate'] = pd.to_datetime(df['ShipDate'])

    df['OrderDate'] = df['OrderDate'].dt.normalize()  # Keeps datetime64 format with time 00:00:00
    df['ShipDate'] = df['ShipDate'].dt.normalize()

    # 1. Shipping delay in days
    df['ShippingDelay'] = df['DaystoShipActual'] - df['DaystoShipScheduled']

    df['IsShippingDelayed'] = df['ShippingDelay'].apply(lambda x: 'Yes' if x > 0 else 'No')

    df = df.drop(columns=['ShippingDelay'])

    # 2.Remove the columns 'DaystoShipActual' and 'DaystoShipScheduled'
    df = df.drop(columns=['DaystoShipActual', 'DaystoShipScheduled'])

    # 3. Calculate the difference in days between OrderDate and ShipDate
    df['DaysBetweenOrderAndShip'] = (df['ShipDate'] - df['OrderDate']).dt.days

    # 4. Calculate the difference in months between OrderDate and ShipDate
    df['MonthsBetweenOrderAndShip'] = ((df['ShipDate'].dt.year - df['OrderDate'].dt.year) * 12) + (df['ShipDate'].dt.month - df['OrderDate'].dt.month)

    # 5.Remove the columns OrderDate and ShipDate
    #df = df.drop(columns=['OrderDate', 'ShipDate'])

    return df
```

```python
df = feature_eng(df)
```
